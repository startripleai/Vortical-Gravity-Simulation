```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_vortical_forces():
    # 1. Physics Parameters
    r = np.linspace(0.1, 2.5, 1000)
    G_M = 1.0     
    Zeta = 1.0   
    rv = 1.0
    rh = 2.0      # Event Horizon

    
    # 2. Force Calculations
    Fg = -G_M / r**2
    Fvr = Zeta / r**3
    F_total = Fg + Fvr
    
    # 3. Stability Radius (Where F_net = 0)
    rv = Zeta / G_M
    
    # 4. Visualization
    # plt.figure(figsize=(10, 7), dpi=300)
    plt.figure(figsize=(10, 6), dpi=100)
    
    # Plotting Curves
    plt.plot(r, Fg, 'r-', label=r'Inward Gravity ($F_g \propto 1/r^2$)', linewidth=2.5)
    plt.plot(r, Fvr, 'b-', label=r'Vortical Repulsion ($F_{vr} \propto 1/r^3$)', linewidth=2.5)
    plt.plot(r, F_total, 'k--', label='Net Force ($F_{total}$)', linewidth=1.5, alpha=0.8)
    
    # --- FIXED SCATTER LINE ---
    # We pass rv as X and 0 as Y
    # plt.scatter(rv, 0, color='black', s=100, zorder=5, label=f'Stability Radius ($r_v={rv}$)')
    # 곡선 및 포인트
    # plt.plot(r, F_total, 'purple', linewidth=2, label='Net Force ($F_{total}$)')
    plt.scatter(rv, 0, color='black', s=80, zorder=5, label='Transition Point (r=1.0)')
    # ---------------------------
    
    # 5. Styling and Shading
    # plt.axhline(0, color='black', linewidth=1)
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(2.0, color='gray', linestyle=':', label='Event Horizon (r=2.0)') # horizon mark
    
    # plt.fill_between(r, 0, F_net, where=(r < rv), color='blue', alpha=0.1, label='Vortical Void (Repulsion Zone)')
    # plt.fill_between(r, 0, F_net, where=(r > rv), color='red', alpha=0.05, label='Torus (Compression Zone)')
    plt.fill_between(r, 0, F_total, where=(r <= rv), color='blue', alpha=0.15, label='Vortical Void (0 < r < 1)')
    plt.fill_between(r, 0, F_total, where=((r > rv) & (r <= 2.0)), color='red', alpha=0.1, label='Resonant Torus (1 < r < 2)')

    
    # plt.title('Asymptotic Analysis: Force Balance at the Lattice Saturation Limit', fontsize=14, fontweight='bold')
    plt.xlabel('Radial Distance (r)', fontsize=12)
    plt.ylabel('Force Magnitude', fontsize=12)
    # plt.ylim(-15, 15)
    # plt.xlim(0.1, 1.0)
    # plt.legend(frameon=True, loc='upper right')
    # plt.grid(True, alpha=0.3)
    # 스타일링
    # plt.title('Vortex Model: Force Equilibrium at r=1.0', fontsize=14)
    plt.title('Asymptotic Force Analysis: The Vortical Void (r<1) and Torus (1<r<2)', fontsize=13)
    # plt.xlabel('Radial Distance (r)')
    # plt.ylabel('Force Magnitude')
    plt.ylim(-2, 2) # Scale so the numbers are clearly visible
    plt.xlim(0.1, 2.2)
    plt.legend(frameon=True, loc='upper right')
    plt.grid(True, alpha=0.3)


    plt.savefig('vortical_force_analysis.png', bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    simulate_vortical_forces()
```
